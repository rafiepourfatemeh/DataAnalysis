```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([0, 10])
ypoints = np.array([20, 100])
#o+xX*.,sdpPHhv<>1342_|  linestyle
#- :-- .-
#rgbcmywk  color
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'sans serif','color':'darkred','size':15}
plt.grid()
plt.plot(xpoints, ypoints,'o-r',ms=20,mec = 'g',mfc = 'y',linewidth=5)
plt.title("mft",loc='left')
plt.xlabel("mft grade",fontdict=font1)
plt.ylabel("mft number",fontdict=font2)
plt.show()
```
