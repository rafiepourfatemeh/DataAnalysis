```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
lesson=['python','ML','java script','react','node js']
number=[1000,700,550,200,150]
```

```python
plt.plot(lesson,number)
plt.show()
```

```python

plt.scatter(lesson,number)
plt.show()
```

```python
plt.hist(number,bins='auto')
#number=[1000,700,550,200,150]
```

```python
plt.pie(number,labels=lesson)
plt.show()
```

```python
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel('lesson')
plt.ylabel('number')
plt.yticks([1000,700,550,200,150],['1000m','700m','550m','200m','150m'])
plt.xticks(rotation=90)
plt.margins(0.7)
plt.text(0,1000,'python',fontsize='10')
plt.text(1,700,'ML',fontsize='10')
plt.plot(lesson,number)
plt.show()
```

```python
lesson=['python','ML','java script','react','node js']
mft=[900,600,450,100,50]
sharif=[1000,700,550,200,150]
```

```python
plt.title('mft&sharif')
plt.xlabel('lesson')
plt.ylabel('number')
plt.plot(lesson,mft,ls='--',marker='+',mew=3,lw=3)
plt.plot(lesson,sharif,ls='-',marker='*',lw=3,mew=2)
plt.legend(['mft','sharif'],loc='best')
plt.grid()
plt.annotate('mftpython',xytext=(0.5,500),xy=(0,900),arrowprops={'facecolor':'red','width':2},fontsize=10)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(lesson,mft)
plt.title('mft')
plt.subplot(1,2,2)
plt.plot(lesson,sharif)
plt.title('sharif')
plt.show()
```
